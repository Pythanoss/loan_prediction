```python
from tqdm import tqdm
class makeit:
    #def __init__(self):
        #print('initializing')
    def fit(self,df,x,y,name):
        self.dic={}
        a=[]
        for i in list(x.value_counts().keys()):
            b=[]
            for j in list(y.value_counts().keys()):
                temp=len(df[(x==i) & (y==j)])
                b.append(temp)
            b=np.array(b)
            self.dic[i]=b/(b.sum())
            #print(self.dic)
    
    def transform(self,df,x,name):
        df[str(name)+'_'+str(0)]=None
        df[str(name)+'_'+str(1)]=None
        for i in tqdm(range(len(df))):
            for j in list(x.value_counts().keys()):
                if df[name][i]==j:
                    try:
                        df[str(name)+'_'+str(1)][i]=self.dic[j][0]
                        df[str(name)+'_'+str(0)][i]=self.dic[j][1]
                    except:
                        df[str(name)+'_'+str(1)][i]=0.5
                        df[str(name)+'_'+str(0)][i]=0.5
                
```

```python
d=makeit()
d.fit(train,train['Dependents'],train['Loan_Status'],'Dependents')
d.transform(train,train['Dependents'],'Dependents')
d.transform(test,test['Dependents'],'Dependents')
```

```python
d=makeit()
d.fit(train,train['Loan_Amount_Term'],train['Loan_Status'],'Loan_Amount_Term')
d.transform(train,train['Loan_Amount_Term'],'Loan_Amount_Term')
d.transform(test,test['Loan_Amount_Term'],'Loan_Amount_Term')
```

```python
train['Gender']=train['Gender'].replace({'Male':0,'Female':1})
test['Gender']=test['Gender'].replace({'Male':0,'Female':1})
```
